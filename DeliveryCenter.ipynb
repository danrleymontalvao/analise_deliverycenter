```python
#importando os pacotes para a análise
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importando os arquivos csv
order= pd.read_csv('orders.csv')
store= pd.read_csv('stores.csv',encoding='latin1')
```

```python
order
```

```python
store
```

```python
#unindo os dois dataframes e exibindo o resultado
df=pd.merge(order,store, how='left',on='store_id')
df
```

```python
#exibindo as lojas por ordem decrescente de vendas
df.store_id.value_counts()
```

```python
#plot do gráfico da porcentagem de vendas por tipo de produto
pie=df.store_segment.value_counts()
labels= pie.index
sizes= pie.values
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.title('Gráfico da porcentagem de vendas por tipo de produto')
plt.gcf().set_size_inches(6, 6)
ax1.axis('equal')
plt.show()
```

```python
#plot do gráfico da porcentagem de pedidos finalizados e cancelados
pie2=df.order_status.value_counts()
labels= pie2.index
sizes= pie2.values
c=['mediumseagreen','red']
fig2, ax2 = plt.subplots()
ax2.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=0,colors=c)
plt.title('Gráfico com a porcentagem de pedidos finalizados e cancelados da plataforma')
ax2.axis('equal')
plt.gcf().set_size_inches(6, 6)
plt.show()
```

```python
#importando o arquivo csv com dados de pagamentos da plataforma
pay= pd.read_csv('payments.csv')
pay
```

```python
#criação de uma tabela dinâmica com o somatório de valores por meio de pagamento
df_pay= pd.pivot_table(pay,columns=['payment_method'],values=['payment_amount'],aggfunc='sum')
df_pay= df_pay.transpose()
df_pay.sort_values('payment_amount',ascending=False)
```

```python
#plot do gráfico com a porcentagem de cada forma de pagamento
x1=df_pay.index
y1=df_pay['payment_amount']
plt.pie(y1,labels=x1,autopct='%1.1f%%')
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
plt.gcf().set_size_inches(20, 18)
fig.gca().add_artist(centre_circle)
plt.title('Gráfico das formas de pagamento de acordo o valor')
plt.show()
```

```python
#transformando os valores da coluna order_moment_finished em data e colocando-os na 
#coluna date
df['date']=pd.to_datetime(df['order_moment_finished']).dt.date
```

```python
#fatiando o dataframe com as informações da loja com mais vendas na plataforma
best=df[df["store_id"] == 53]
best
```

```python
#criando um coluna com o valor total pago pelo cliente, o que inclui:
#valor do produto + taxas
best['total']=best['order_amount']+best['order_delivery_fee']+best['order_delivery_cost']
```

```python
#criando uma tabela dinâmica com a quantidade de vendas por data
df2=pd.pivot_table(best,columns=['date'],values=['total'],aggfunc='sum')
df2=df2.transpose()
df2
```

```python
#resumo estatístico das vendas
best['total'].describe()
```

```python
#plot do gráfico com a linha do tempo das vendas da loja PAPA SUCIS
plt.plot(df2.index, df2.values)
plt.scatter(df2.index, df2.values, label = "Valor da venda", color = "r", marker = ".", s = 100)
plt.title('Vendas da loja PAPA SUCIS entre os meses de janeiro e maio de 2021')
plt.xlabel('Dia')
plt.ylabel('Valor da venda em R$')
plt.legend()
plt.gcf().set_size_inches(18, 12)
plt.show()
```

```python
#transformando as datas em dias da semana e colocando-as numa outra coluna
best['day']=best['date']
for i in best.index:
    best['day'][i]=pd.Timestamp(best['date'][i]).day_name()
```

```python
#criando uma tabela dinâmica com o valor de vendas por dia da semana
df3=pd.pivot_table(best,columns=['day'],values=['total'],aggfunc='sum')
df3=df3.transpose()
df3.sort_values('total',ascending=False)
```

```python
#plot do gráfico de vendas por dia da semana
x3=['Domingo','Segunda','Terça','Quarta','Quinta','Sexta','Sábado']
y3=[163918.68,144241.41,152959.90,155450.57,158312.90,163753.02,154048.14]
plt.bar(x3,y3,color='crimson')
plt.gcf().set_size_inches(12,8)
plt.xlabel('Dia da Semana')
plt.ylabel('Valor das vendas em R$')
plt.title('Gráfico da quantidade de vendas por dia da semana')
plt.show()
```
