```python
#importando os pacotes para a an√°lise
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importando os arquivos csv
order= pd.read_csv('orders.csv')
store= pd.read_csv('stores.csv',encoding='latin1')
```

```python
order
```

```python
store
```

```python
#unindo os dois dataframes e exibindo o resultado
df=pd.merge(order,store, how='left',on='store_id')
df
```

```python
#exibindo as lojas por ordem decrescente de vendas
df.store_id.value_counts()
```

```python
#plot do gr√°fico da porcentagem de vendas por tipo de produto
pie=df.store_segment.value_counts()
labels= pie.index
sizes= pie.values
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.title('Gr√°fico da porcentagem de vendas por tipo de produto')
plt.gcf().set_size_inches(6, 6)
ax1.axis('equal')
plt.show()
```

```python
#plot do gr√°fico da porcentagem de pedidos finalizados e cancelados
pie2=df.order_status.value_counts()
labels= pie2.index
sizes= pie2.values
c=['mediumseagreen','red']
fig2, ax2 = plt.subplots()
ax2.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=0,colors=c)
plt.title('Gr√°fico com a porcentagem de pedidos finalizados e cancelados da plataforma')
ax2.axis('equal')
plt.gcf().set_size_inches(6, 6)
plt.show()
```

```python
#importando o arquivo csv com dados de pagamentos da plataforma
pay= pd.read_csv('payments.csv')
pay
```

```python
#cria√ß√£o de uma tabela din√¢mica com o somat√≥rio de valores por meio de pagamento
df_pay= pd.pivot_table(pay,columns=['payment_method'],values=['payment_amount'],aggfunc='sum')
df_pay= df_pay.transpose()
df_pay.sort_values('payment_amount',ascending=False)
```

```python
#plot do gr√°fico com a porcentagem de cada forma de pagamento
x1=df_pay.index
y1=df_pay['payment_amount']
plt.pie(y1,labels=x1,autopct='%1.1f%%')
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
plt.gcf().set_size_inches(20, 18)
fig.gca().add_artist(centre_circle)
plt.title('Gr√°fico das formas de pagamento de acordo o valor')
plt.show()
```

```python
#transformando os valores da coluna order_moment_finished em data e colocando-os na 
#coluna date
df['date']=pd.to_datetime(df['order_moment_finished']).dt.date
```

```python
#fatiando o dataframe com as informa√ß√µes da loja com mais vendas na plataforma
best=df[df["store_id"] == 53]
best
```

```python
#criando um coluna com o valor total pago pelo cliente, o que inclui:
#valor do produto + taxas
best['total']=best['order_amount']+best['order_delivery_fee']+best['order_delivery_cost']
```

```python
#criando uma tabela din√¢mica com a quantidade de vendas por data
df2=pd.pivot_table(best,columns=['date'],values=['total'],aggfunc='sum')
df2=df2.transpose()
df2
```

```python
#resumo estat√≠stico das vendas
best['total'].describe()
```

```python
#plot do gr√°fico com a linha do tempo das vendas da loja PAPA SUCIS
plt.plot(df2.index, df2.values)
plt.scatter(df2.index, df2.values, label = "Valor da venda", color = "r", marker = ".", s = 100)
plt.title('Vendas da loja PAPA SUCIS entre os meses de janeiro e maio de 2021')
plt.xlabel('Dia')
plt.ylabel('Valor da venda em R$')
plt.legend()
plt.gcf().set_size_inches(18, 12)
plt.show()
```

```python
#transformando as datas em dias da semana e colocando-as numa outra coluna
best['day']=best['date']
for i in best.index:
    best['day'][i]=pd.Timestamp(best['date'][i]).day_name()
```

```python
#criando uma tabela din√¢mica com o valor de vendas por dia da semana
df3=pd.pivot_table(best,columns=['day'],values=['total'],aggfunc='sum')
df3=df3.transpose()
df3.sort_values('total',ascending=False)
```
